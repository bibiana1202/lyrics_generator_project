```python
repeat_removed_titles = []        #반복되는 글자 지운 노래들을 담아줄 리스트 선언


for i in range(len(removed_title_ls2)) :
  repeat_removed_title = []                                                   # 반복된 글자를 지운 노래하나를 담아 둘 리스트 선언
  repeat_removed_title.append(removed_title_ls2[i][0])                                 #첫 번째 토큰을 넣어둠
  for j in range(1, len(removed_title_ls2[i])):          
    if removed_title_ls2[i][j] != repeat_removed_title[-1] :          #2번째 토큰부터 리스트의 마지막 인덱스값과 비교해가면서(마지막값과 같으면 반복인것)
      repeat_removed_title.append(removed_title_ls2[i][j])                 #노래하나 담을 리스트에 추가    
  repeat_removed_titles.append(repeat_removed_title)                        #전체 노래 담을 리스트에 추가

```
